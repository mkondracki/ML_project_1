# Standardize data

We need to replace the missing data with smart values

1. Separate class and reg data 

```python
x_tr_class = x_tr[:, 22]
x_tr_reg = np.delete(x_tr, 22, axis=1)

x_te_class = x_te[:, 22]
x_te_reg = np.delete(x_te, 22, axis=1)
```

2. replace outliers by median 

```python
# Not standardize but value of nan replace 
x_tr_reg = replace_nan_by_median(x_tr_reg)
x_te_reg = replace_nan_by_median(x_te_reg)
```

3. standardize matrix 

```python
x_tr_st = standardize(x_tr_reg)
x_te_st = standardize(x_te_reg)
```

# Multiply features and apply low variance 

```python
lambda_ = 10**(-7)
max_iters = 1000
gamma = 10**(-1)
threshold = 0.7

#standardize the x and initialze w 
initial_w = np.full(x_tr_st.shape[1], 0.1)

w, loss = reg_logistic_regression(y_tr, x_tr_st, lambda_, initial_w, max_iters, gamma)
print("Not doing anything to our features give us ", loss)

#standardize the x and initialze w 
x_tr_temp = multiply_features(x_tr_reg)
x_tr_m = standardize(x_tr_temp)
initial_w = np.full(x_tr_m.shape[1], 0.1)

w, loss = reg_logistic_regression(y_tr, x_tr_m, lambda_, initial_w, max_iters, gamma)
print("Feature expansion we obtain a loss of", loss)

#standardize the x and initialze w 
var, x_var = remove_low_variance(x_tr_reg, threshold)
x_tr_temp = multiply_features(x_var)
x_tr_m_var = standardize(x_tr_temp)
initial_w = np.full(x_tr_m_var.shape[1], 0.1)

w, loss = reg_logistic_regression(y_tr, x_tr_m_var, lambda_, initial_w, max_iters, gamma)
print("By combining low variance feature selection and feature expansion we obtain a loss of", loss)
```

Not doing anything to our features give us  0.5015956236731489

 by combining low variance feature selection and feature expansion we obtain a loss of 0.4845835755702653
 
 So we can conclude that the feature expansion using multiply in between the columns improve drastically our result. removing low variance doesn't work
